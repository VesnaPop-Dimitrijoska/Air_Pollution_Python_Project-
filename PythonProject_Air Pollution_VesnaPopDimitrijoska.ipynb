## Introduction

The city of Skopje is well known for its struggles with poor air quality, especially in the winter months. The main pollutants are CO, NO2, ozone, and particulate matter of different dimensions.

Monitoring stations for measuring the concentrations of each of the main pollutants have been installed in different locations around the city.

## Import Libraries
Before we begin, let us import the libraries that we will be using (Pandas, matplotlib and seaborn). Any additional libraries that we decide to import we will add in this cell.

```python
# Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## The Dataset
The following data has been taken from kaggle (https://www.kaggle.com/datasets/cokastefan/pm10-pollution-data-in-skopje-from-2008-to-2018) where data from the monitoring stations (https://air.moepp.gov.mk/?page_id=175) for the period from Janaury 2007 to December 2018 has been compiled in a single dataset. To make our task more managable, we will only be looking at data for the time period of five years, from January 2012 to December 2016.

## Import data
The data is contained in the directory CSVs in the files (CO.csv, NO2.csv, O3.csv, PM10.csv, and PM25.csv). Let's import the data from the CSV files as separate pandas dataframes.

```python
# Read files into dataframes

df_CO   = pd.read_csv('CSVs/CO.csv')
df_NO2  = pd.read_csv('CSVs/NO2.csv')
df_O3   = pd.read_csv('CSVs/O3.csv')
df_PM10 = pd.read_csv('CSVs/PM10.csv')
df_PM25 = pd.read_csv('CSVs/PM25.csv')
```

## Review Data
Let's display each of the dataframes and look at the data types in order to get a sense of the data. Check that the data types are accurate.

## Data cleaning and restructuring
Because the mobile column is from a mobile monitoring station located in the municipality of Butel and occasionally displaced, we can drop that column.

We can also rename the column time to datetime and create additional separate time and date columns. 

We can add a column in each of the dataframes indicating the type of pollutant and concatenate the dataframes.

```python
# 1) Adding a 'Pollutant' column

df_CO  ['Pollutant'] = 'CO'
df_NO2 ['Pollutant'] = 'NO2'
df_O3  ['Pollutant'] = 'O3'
df_PM10['Pollutant'] = 'PM10'
df_PM25['Pollutant'] = 'PM25' 
```

```python
# 2) Concatenation

df = pd.concat([df_CO, df_NO2, df_O3, df_PM10, df_PM25], ignore_index = True)
```

```python
# 3) Deleting columns

del df['Unnamed: 0.1'], df['Unnamed: 0'], df['Mobile']
```

```python
# 4) Renaming columns

df.rename(columns={'time': 'Datetime'}, inplace=True)
```

```python
# 5) Deleting duplicates

df.drop_duplicates(inplace=True) 
# df.duplicated().to_excel('DuplicatesAnalysis.xlsx') 
```

```python
# 6) Conversion to appropriate data types

df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
# 7) Creating time columns

df['Date']  = df['Datetime'].dt.date
df['Time']  = df['Datetime'].dt.time
df['Year']  = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Day']   = df['Datetime'].dt.day
df['Hour']  = df['Datetime'].dt.hour
```

```python
# 8) Creation of additional lists

locations = ['Centar', 'GaziBaba', 'Karpos', 'Lisice', 'Miladinovci', 'Rektorat']
pollutants = ['CO', 'NO2', 'O3', 'PM10', 'PM25']
```

```python
# 9) Calculating % of NaN values

def nan_percentages(df):
    
    for col in df.columns:  
        print(f'{col} : {round((df[col].isna().sum() / len(df[col])) * 100, 2)}%')

nan_percentages(df)
```

# Question 1
What location in Skopje has the highest average pollution for each of the pollutants? .......... (2 points)

```python
# Solution

# Group by pollutant and calculate mean for each location
df_average_Q1 = df.groupby(['Pollutant'])[locations].mean()

print('\nМаксимално просечно загадување на сите мерни локации во Скопје со: CO, NO2, O3, PM10 и PM25 е дадено во табелата:')
df_average_Q1
```

```python
# Get the location with the highest average pollution for each pollutant. Function returns index of axis maximum.
max_locations_Q1 = df_average_Q1.idxmax(axis=1)

# Get the maximum average pollution for each pollutant
max_values_Q1 = df_average_Q1.max(axis=1)

# Combine the results into a single dataframe
result_Q1 = pd.DataFrame({'Location': max_locations_Q1, 'Max Value': max_values_Q1})

print('\nЗАКЛУЧОК:\n\nЛокација со максимално просечно загадување во Скопје со: CO, NO2, О3, PM10 и PM25 е дадена во табелата:')
result_Q1
```

# Question 2
For every year, for each of the pollutants, on what time stamp is the maximum pollution for every location? What time of the year seems to have the most pollution in Skopje? .......... (2 points)

```python
# Solution - Location: Centar

# Filter the dataframe to keep only the rows with the 'Centar' location
centar_df = df[['Year', 'Pollutant', 'Centar', 'Datetime']]
centar_df = centar_df[centar_df['Centar'].notnull()]

# Group the filtered dataframe by year and pollutant, and calculate the maximum pollution value for each group
grouped = centar_df.groupby(['Year', 'Pollutant'])['Centar'].max().reset_index()
grouped = grouped.rename(columns={'Centar': 'Max Pollution'})

# Merge the maximum pollution values back into the original dataframe based on the year, pollutant, and max pollution value
merged = pd.merge(centar_df, grouped, on=['Year', 'Pollutant'])

# Filter the merged dataframe to keep only the rows with the maximum pollution value for the 'Centar' location
filtered = merged[merged['Centar'] == merged['Max Pollution']]
result_Q2_Centar = filtered[['Year', 'Pollutant', 'Centar', 'Datetime']]

print('\nЗАКЛУЧОК:\n\nСогласно резултатите за локација Центар, максимални измерени вредности на CO, NO2, PM10 и PM25 има во текот на зимските месеци, додека на О3 во текот на летните месеци.')
print('\nТабеларен приказ со максимални вредности за локација Центар:')
result_Q2_Centar
```

## Question 7
On how many days in each year, was the value of PM10 measured on at least one location worse than moderate? .......... (3 points)

```python
# Solution

df_PM10 = df[df['Pollutant'] == 'PM10'][['Year', 'Date', 'Datetime', 'Centar', 'GaziBaba', 'Karpos', 'Lisice', 'Miladinovci', 'Rektorat']]

df_PM10_cond = df_PM10[(df_PM10['Centar'] >= 155) 
              | (df_PM10['GaziBaba'] >= 155) 
              | (df_PM10['Karpos'] >= 155) 
              | (df_PM10['Lisice'] >= 155)
              | (df_PM10['Miladinovci'] >= 155)
              | (df_PM10['Rektorat'] >= 155)][['Year', 'Date', 'Centar', 'GaziBaba', 'Karpos', 'Lisice', 'Miladinovci', 'Rektorat']]

df_PM10_daily = df_PM10_cond.groupby(['Year', 'Date']).size()
df_PM10_yearly = df_PM10_daily.groupby('Year').count().reset_index(name='Days with unhealthy PM10 levels')

print('\nТабеларен приказ на денови во годината во кој имаме барем едно надминување на PM10 честички поголемо од 155 мерни единици:')
df_PM10_yearly
```

## Question 8
Make a bar plot showing the average number of days per year when the measured value for PM10 was worse than moderate on at least one location.  .......... (3 points)

```python
# Solution

df_PM10_avg = df_PM10_yearly.mean().values[1]

fig = plt.figure(figsize=(7, 4))
sns.barplot(data=df_PM10_yearly, x='Year', y='Days with unhealthy PM10 levels', alpha=0.7, width = 0.6)
plt.axhline(y=df_PM10_avg, color='red', linestyle='--', label='Average number of days')
plt.legend(title='Year', loc='lower center', bbox_to_anchor=(1.35, 0.5), ncol=1)
plt.title('Days with unhealthy PM10 pollution levels') 
plt.show()
```
